```python
import csv
import pandas as pd
import numpy as np
```

```python
x = []
y = []
with open('dataset.csv','r') as f:
    reader = csv.reader(f)
    next(reader, None)
    for i in reader:
        
        x.append(i[1:])
        y.append(i[0])

org_x = x
org_y = y
x = np.array(x).flatten()
x = np.unique(x)
x = np.delete(x,0) 
y = np.array(y)
y = np.unique(y)
```

```python
y
```

```python
new_data = []
i = 0
for row in org_x:
    z = []
    for sn in x:
        if sn in row:
            z.append(1)
        else:
            z.append(0)
    dn = org_y[i]
    dc = np.where(y == dn)[0][0]
    z.append(dn)
    z.append(dc)
    new_data.append(z)
    i+=1
```

```python
dn = 'GERD'
dc = np.where(y == dn)[0]
dc[0]
```

```python
x = np.append(x,'Disease')
x = np.append(x,'Disease Class')
x
```

```python
with open('new_data.csv','w') as f:
    writer = csv.writer(f)
    writer.writerow(x)
    writer.writerows(new_data)
```
